```python
import requests, time, random
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Firefox( executable_path = "driver/geckodriver.exe")
driver.get('https://www.linkedin.com/uas/login')
#browser = webdriver.Chrome('driver/chromedriver.exe')
#browser.get('https://www.linkedin.com/uas/login')
file = open('config.txt')
lines = file.readlines()
username = lines[0]
password = lines[1]


elementID = driver.find_element_by_xpath('//*[@id="username"]')
elementID.send_keys(username)

elementID = driver.find_element_by_xpath('//*[@id="password"]')
elementID.send_keys(password)

elementID.submit()
```

```python
#link='https://www.linkedin.com/in/sachin-gaykhe-1476128a/'
link = 'https://www.linkedin.com/in/ankit-nayase/'
driver.get(link)
```

```python
SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

for i in range(3):
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
src = driver.page_source
soup = BeautifulSoup(src, 'lxml')
soup
```

```python
name_div = soup.find('div', {'class': 'flex-1 mr5'})
name_div
```

```python
name_loc = name_div.find_all('ul')
name = name_loc[0].find('li').get_text().strip()
name
```

```python
loc = name_loc[1].find('li').get_text().strip()
loc
```

```python
profile_title = name_div.find('h2').get_text().strip()
profile_title
```

```python
connection = name_loc[1].find_all('li')
connection = connection[1].get_text().strip()
connection
```

```python
info = []
info.append(link)
info.append(name)
info.append(profile_title)
info.append(loc)
info.append(connection)
info
```
